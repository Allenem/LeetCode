## 239. 滑动窗口最大值

已解答
困难
相关标签

相关企业
提示

给你一个整数数组 nums，有一个大小为 k 的滑动窗口从数组的最左侧移动到数组的最右侧。你只可以看到在滑动窗口内的 k 个数字。滑动窗口每次只向右移动一位。

返回 滑动窗口中的最大值 。

 

示例 1：

```
输入：nums = [1,3,-1,-3,5,3,6,7], k = 3
输出：[3,3,5,5,6,7]
解释：
滑动窗口的位置                最大值
---------------               -----
[1  3  -1] -3  5  3  6  7       3
 1 [3  -1  -3] 5  3  6  7       3
 1  3 [-1  -3  5] 3  6  7       5
 1  3  -1 [-3  5  3] 6  7       5
 1  3  -1  -3 [5  3  6] 7       6
 1  3  -1  -3  5 [3  6  7]      7
```

示例 2：
```
输入：nums = [1], k = 1
输出：[1]
```

提示：
```
1 <= nums.length <= 105
-104 <= nums[i] <= 104
1 <= k <= nums.length
```

````python
from typing import List

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        if not nums or not k: return []
        if len(nums) == 1: return [nums[0]]
        # 永远保持maxidx第一位所对应的num是窗口中最大的，递减到i
        maxidx, ret = [], []
        for i in range(len(nums)):
            # 判断第一个索引是否过界，因为第一个最远
            if maxidx and maxidx[0] == i-k:
                maxidx.pop(0)
            # 如果当前num比maxidx最后一个索引对应的num大，说明maxidx最后一个索引对应的num不可能是窗口最大了，弹出，
            # 直到maxidx最后一个索引对应的num比当前num大，或者maxidx空了
            while maxidx and nums[maxidx[-1]] < nums[i]:
                maxidx.pop()
            maxidx.append(i)
            if i >= k-1:
                ret.append(nums[maxidx[0]])

            print(f'i={i}, maxidx={maxidx}, ret={ret}')
        return ret
    
s = Solution()
print(s.maxSlidingWindow([1,3,-1,-3,5,3,6,7], 3))
````

示例 1:
```
输入: s = "abcabcbb"
输出: 3 
解释: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。注意 "bca" 和 "cab" 也是正确答案。
```
示例 2:
```
输入: s = "bbbbb"
输出: 1
解释: 因为无重复字符的最长子串是 "b"，所以其长度为 1。
```
示例 3:
```
输入: s = "pwwkew"
输出: 3
解释: 因为无重复字符的最长子串是 "wke"，所以其长度为 3。
     请注意，你的答案必须是 子串 的长度，"pwke" 是一个子序列，不是子串。
```

提示：
```
0 <= s.length <= 5 * 104
s 由英文字母、数字、符号和空格组成
```

````python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        d, i, ret = {}, -1, 0
        for j in range(len(s)):
            if s[j] in d:
                i = max(d[s[j]], i) # 更新左窗口（开始无重复的前一个索引，方便直接减出片段长度）
            d[s[j]] = j
            ret = max(j-i, ret)
        return ret

s = Solution()
print(s.lengthOfLongestSubstring('abcabcbb'))
print(s.lengthOfLongestSubstring('bbbbb'))
print(s.lengthOfLongestSubstring('pwwkew'))
````

## 438. 找到字符串中所有字母异位词

已解答

给定两个字符串 s 和 p，找到 s 中所有 p 的 异位词 （字母异位词是通过重新排列不同单词或短语的字母而形成的单词或短语，并使用所有原字母一次。） 的子串，返回这些子串的起始索引。不考虑答案输出的顺序。

示例 1:
```
输入: s = "cbaebabacd", p = "abc"
输出: [0,6]
解释:
起始索引等于 0 的子串是 "cba", 它是 "abc" 的异位词。
起始索引等于 6 的子串是 "bac", 它是 "abc" 的异位词。
```
示例 2:
```
输入: s = "abab", p = "ab"
输出: [0,1,2]
解释:
起始索引等于 0 的子串是 "ab", 它是 "ab" 的异位词。
起始索引等于 1 的子串是 "ba", 它是 "ab" 的异位词。
起始索引等于 2 的子串是 "ab", 它是 "ab" 的异位词。
```

提示:
```
1 <= s.length, p.length <= 3 * 104
s 和 p 仅包含小写字母
```

````python
from collections import Counter

class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        cntp = Counter(p)
        cnts = Counter()
        ans = []
        for r, c in enumerate(s):
            cnts[c] += 1
            l = r - len(p) + 1
            if l < 0: continue
            if cnts == cntp: ans.append(l)
            cnts[s[l]] -= 1
        return ans
    
s = Solution()
print(s.findAnagrams('cbaebabacd', 'abc'))
print(s.findAnagrams('abab', 'ab'))
````

## 76. 最小覆盖子串

已解答
困难
相关标签
premium lock icon
相关企业
提示

给定两个字符串 s 和 t，长度分别是 m 和 n，返回 s 中的 最短窗口 子串 (子字符串 是字符串中连续的 非空 字符序列。)，使得该子串包含 t 中的每一个字符（包括重复字符）。如果没有这样的子串，返回空字符串 ""。

测试用例保证答案唯一。

示例 1：
```
输入：s = "ADOBECODEBANC", t = "ABC"
输出："BANC"
解释：最小覆盖子串 "BANC" 包含来自字符串 t 的 'A'、'B' 和 'C'。
```
示例 2：
```
输入：s = "a", t = "a"
输出："a"
解释：整个字符串 s 是最小覆盖子串。
```
示例 3:
```
输入: s = "a", t = "aa"
输出: ""
解释: t 中两个字符 'a' 均应包含在 s 的子串中，
因此没有符合条件的子字符串，返回空字符串。
```

提示：
```
m == s.length
n == t.length
1 <= m, n <= 105
s 和 t 由英文字母组成
```

进阶：你能设计一个在 O(m + n) 时间内解决此问题的算法吗？



````python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        ht = Counter(t)
        hs = Counter()
        ans = ''
        lent = len(t)
        l = 0
        effectcnt = 0
        for r, c in enumerate(s):
            hs[s[r]] += 1
            # 如果当前字符在t中，并且当前字符在窗口中的数量不超过t中该字符的数量，说明这个字符对满足条件是有贡献的，effectcnt加1
            if hs[s[r]] <= ht[s[r]]: effectcnt += 1
            # 如果当前窗口长度大于等于t的长度，并且当前窗口中s[l]的数量超过t中s[l]的数量，说明s[l]对满足条件没有贡献了，l右移，effectcnt不变
            while r - l + 1 >= lent and hs[s[l]] > ht[s[l]]:
                hs[s[l]] -= 1
                l += 1
            # 如果当前窗口长度大于等于t的长度，并且effectcnt等于t的长度，说明当前窗口满足条件了，更新答案
            if effectcnt == lent:
                # 如果当前没有答案，或者当前窗口长度比之前的答案更短，更新答案
                if not ans or r-l+1 < len(ans):
                    ans = s[l:r+1]
        return ans

s = Solution()
print(s.minWindow('ADOBECODEBANC', 'ABC'))
print(s.minWindow('a', 'a'))
print(s.minWindow('a', 'aa'))
````

## 209. 长度最小的子数组

给定一个含有 n 个正整数的数组和一个正整数 target 。

找出该数组中满足其总和大于等于 target 的长度最小的 子数组 (子数组 是数组中连续的 非空 元素序列。) [numsl, numsl+1, ..., numsr-1, numsr] ，并返回其长度。如果不存在符合条件的子数组，返回 0 。

示例 1：
```
输入：target = 7, nums = [2,3,1,2,4,3]
输出：2
解释：子数组 [4,3] 是该条件下的长度最小的子数组。
```
示例 2：
```
输入：target = 4, nums = [1,4,4]
输出：1
```
示例 3：
```
输入：target = 11, nums = [1,1,1,1,1,1,1,1]
输出：0
```

提示：
```
1 <= target <= 109
1 <= nums.length <= 105
1 <= nums[i] <= 104
```

进阶：

如果你已经实现 O(n) 时间复杂度的解法, 请尝试设计一个 O(n log(n)) 时间复杂度的解法。

````python
# 解法一：滑窗
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        cursum, l, n = 0, 0, len(nums)
        ans = n+1
        for r in range(n):
            cursum += nums[r]
            while cursum >= target:
                ans = min(ans, r-l+1)
                cursum -= nums[l]
                l += 1
        return ans if ans != n+1 else 0
    
s = Solution()
print(s.minSubArrayLen(7, [2,3,1,2,4,3]))
print(s.minSubArrayLen(4, [1,4,4]))
print(s.minSubArrayLen(11, [1,1,1,1,1,1,1,1]))
````

````python
# 解法二：前缀和+二分
import bisect
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        if not nums: return 0
        n = len(nums)
        ans = n + 1
        presum = [0]
        for i in range(n): presum.append(presum[-1] + nums[i])
        for i in range(1, n + 1):
            # presumi + target = presumj
            # i: left boundary
            # j: right boundary
            presumj = presum[i-1] + target
            j = bisect.bisect_left(presum, presumj)
            if j != len(presum):
                ans = min(ans, j - (i - 1))
        return 0 if ans == n+1 else ans
    
s = Solution()
print(s.minSubArrayLen(7, [2,3,1,2,4,3]))
print(s.minSubArrayLen(4, [1,4,4]))
print(s.minSubArrayLen(11, [1,1,1,1,1,1,1,1]))
````

## 219. 存在重复元素 II

给你一个整数数组 nums 和一个整数 k ，判断数组中是否存在两个 不同的索引 i 和 j ，满足 nums[i] == nums[j] 且 abs(i - j) <= k 。如果存在，返回 true ；否则，返回 false 。

示例 1：
```
输入：nums = [1,2,3,1], k = 3
输出：true
```
示例 2：
```
输入：nums = [1,0,1,1], k = 1
输出：true
```
示例 3：
```
输入：nums = [1,2,3,1,2,3], k = 2
输出：false
```

提示：
```
1 <= nums.length <= 105
-109 <= nums[i] <= 109
0 <= k <= 105
```

````python
class Solution:
    def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
        num1idx = {}
        for j, num in enumerate(nums):
            if num in num1idx and j - num1idx[num] <= k:
                return True
            num1idx[num] = j
        return False
    
s = Solution()
print(s.containsNearbyDuplicate([1,2,3,1], 3))
print(s.containsNearbyDuplicate([1,0,1,1], 1))
print(s.containsNearbyDuplicate([1,2,3,1,2,3], 2))
````
