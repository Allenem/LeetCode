### 2.1 手搓数组

````python
class MyArray:
    def __init__(self, capacity=10):
        self._capacity = capacity
        self._size = 0
        self._data = [None] * capacity
    
    def append(self, value):
        if self._size == self._capacity:
            # 扩容
            self._resize(self._capacity * 2)
        self._data[self._size] = value
        self._size += 1
    
    def _resize(self, new_capacity):
        new_data = [None] * new_capacity
        for i in range(self._size):
            new_data[i] = self._data[i]
        self._data = new_data
        self._capacity = new_capacity
    
    def get(self, index):
        if index < 0 or index >= self._size:
            raise IndexError("Index out of range")
        return self._data[index]
    
    def __getitem__(self, index):
        """支持 arr[index] 语法"""
        if index < 0:
            index = self._size + index  # 支持负数索引
        if index < 0 or index >= self._size:
            raise IndexError("Index out of range")
        return self._data[index]
    
    def __setitem__(self, index, value):
        """支持 arr[index] = value 语法"""
        if index < 0:
            index = self._size + index
        if index < 0 or index >= self._size:
            raise IndexError("Index out of range")
        self._data[index] = value
    
    def __len__(self):
        return self._size
    
    def __str__(self):
        return str([self._data[i] for i in range(self._size)])
    
    def __repr__(self):
        return f"MyArray({[self._data[i] for i in range(self._size)]})"
    
    def __iter__(self):
        """支持迭代"""
        for i in range(self._size):
            yield self._data[i]
    
    def pop(self):
        """删除并返回最后一个元素"""
        if self._size == 0:
            raise IndexError("pop from empty array")
        value = self._data[self._size - 1]
        self._size -= 1
        return value
    
    def insert(self, index, value):
        """在指定位置插入元素"""
        if index < 0:
            index = self._size + index
        if index < 0 or index > self._size:
            raise IndexError("Index out of range")
        
        if self._size == self._capacity:
            self._resize(self._capacity * 2)
        
        # 后移元素
        for i in range(self._size, index, -1):
            self._data[i] = self._data[i-1]
        
        self._data[index] = value
        self._size += 1
    
    def remove(self, value):
        """删除第一个匹配的元素"""
        for i in range(self._size):
            if self._data[i] == value:
                # 前移后续元素
                for j in range(i, self._size - 1):
                    self._data[j] = self._data[j+1]
                self._size -= 1
                return True
        return False

# 测试修复后的数组
def test_my_array():
    print("=== 测试手搓数组 ===")
    
    # 创建数组
    arr = MyArray(5)
    print(f"初始数组: {arr}, 长度: {len(arr)}, 容量: {arr._capacity}")
    
    # 添加元素
    for i in range(1, 8):
        arr.append(i * 10)
    print(f"添加7个元素后: {arr}, 长度: {len(arr)}, 容量: {arr._capacity}")
    
    # 测试 __getitem__
    print(f"\n=== 测试索引访问 ===")
    print(f"arr[0] = {arr[0]}")
    print(f"arr[2] = {arr[2]}")
    print(f"arr[-1] = {arr[-1]}")
    print(f"arr[-2] = {arr[-2]}")
    
    # 测试 __setitem__
    print(f"\n=== 测试修改元素 ===")
    arr[0] = 100
    print(f"修改arr[0]后: {arr}")
    arr[-1] = 999
    print(f"修改arr[-1]后: {arr}")
    
    # 测试迭代
    print(f"\n=== 测试迭代 ===")
    print("遍历数组:", end=" ")
    for item in arr:
        print(item, end=" ")
    print()
    
    # 测试pop
    print(f"\n=== 测试pop ===")
    popped = arr.pop()
    print(f"弹出元素: {popped}, 弹出后数组: {arr}")
    
    # 测试insert
    print(f"\n=== 测试insert ===")
    arr.insert(2, 55)
    print(f"在位置2插入55后: {arr}")
    arr.insert(-1, 88)  # 在倒数第二个位置插入
    print(f"在位置-1插入88后: {arr}")
    
    # 测试remove
    print(f"\n=== 测试remove ===")
    removed = arr.remove(55)
    print(f"删除55: {removed}, 删除后: {arr}")
    
    # 测试异常处理
    print(f"\n=== 测试异常 ===")
    try:
        print(arr[100])
    except IndexError as e:
        print(f"索引100越界: {e}")
    
    try:
        arr[100] = 999
    except IndexError as e:
        print(f"索引999越界: {e}")

# 运行测试
if __name__ == "__main__":
    test_my_array()
    
    # 对比Python原生列表
    print("\n" + "="*50)
    print("对比Python原生列表:")
    
    # 相同操作
    native_list = [10, 20, 30, 40, 50, 60, 70]
    my_arr = MyArray(5)
    for num in native_list:
        my_arr.append(num)
    
    print(f"Python列表: {native_list}")
    print(f"手搓数组: {my_arr}")
    print(f"列表长度: {len(native_list)}, 数组长度: {len(my_arr)}")
    print(f"列表[0]: {native_list[0]}, 数组[0]: {my_arr[0]}")
    print(f"列表[-1]: {native_list[-1]}, 数组[-1]: {my_arr[-1]}")
````

### 2.2 手搓单链表

````python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next
    
    def __str__(self):
        return str(self.value)
    
    def __repr__(self):
        return f"ListNode({self.value})"


class MyLinkedList:
    def __init__(self):
        self.head = None
        self.size = 0
    
    def append(self, value):
        new_node = ListNode(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
        self.size += 1
    
    def get(self, index):
        if index < 0 or index >= self.size:
            # raise IndexError("Index out of range")
            print(f"索引{index}越界: Index out of range")
            return
        current = self.head
        for _ in range(index):
            current = current.next
        return current.value
    
    def __getitem__(self, index):
        """支持 linked_list[index] 语法"""
        return self.get(index)
    
    def __setitem__(self, index, value):
        """支持 linked_list[index] = value 语法"""
        if index < 0 or index >= self.size:
            # raise IndexError("Index out of range")
            print(f"索引{index}越界: Index out of range")
            return
        current = self.head
        for _ in range(index):
            current = current.next
        current.value = value
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        values = []
        current = self.head
        while current:
            values.append(str(current.value))
            current = current.next
        return " -> ".join(values)
    
    def __repr__(self):
        return f"MyLinkedList({str(self)})"


# 测试手搓单链表
def test_my_linked_list():
    print("=== 测试手搓单链表 ===")
    
    linked_list = MyLinkedList()
    print(f"初始链表: {linked_list}, 长度: {len(linked_list)}")
    
    # 添加元素
    for i in range(1, 6):
        linked_list.append(i * 10)
    print(f"添加5个元素后: {linked_list}, 长度: {len(linked_list)}")
    
    # 测试 __getitem__
    print(f"\n=== 测试索引访问 ===")
    print(f"linked_list[0] = {linked_list[0]}")
    print(f"linked_list[2] = {linked_list[2]}")
    print(f"linked_list[-1] = {linked_list[-1]}")  # 不支持负数索引
    
    # 测试 __setitem__
    print(f"\n=== 测试修改元素 ===")
    linked_list[0] = 100
    print(f"修改linked_list[0]后: {linked_list}")
    linked_list[4] = 999
    print(f"修改linked_list[4]后: {linked_list}")
    
    # 测试异常处理
    print(f"\n=== 测试异常 ===")
    try:
        print(linked_list[100])
    except IndexError as e:
        print(f"索引100越界: {e}")
    
    try:
        linked_list[100] = 999
    except IndexError as e:
        print(f"索引100越界: {e}")

# 运行测试
if __name__ == "__main__":
    test_my_linked_list()
````

### 2.3 手搓双链表

````python
class ListNode:
    def __init__(self, value=0, next=None, prev=None):
        self.value = value
        self.next = next
        self.prev = prev

    def __str__(self):
        return str(self.value)
    
    def __repr__(self):
        return f"ListNode({self.value})"
    
class MyLinkedList:
    def __init__(self):
        self.head = None
        self.size = 0
    
    def append(self, value):
        new_node = ListNode(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current
        self.size += 1
    
    def get(self, index):
        if index < 0 or index >= self.size:
            # raise IndexError("Index out of range")
            print(f"索引{index}越界: Index out of range")
            return
        current = self.head
        for _ in range(index):
            current = current.next
        return current.value
    
    def __getitem__(self, index):
        """支持 linked_list[index] 语法"""
        return self.get(index)
    
    def __setitem__(self, index, value):
        """支持 linked_list[index] = value 语法"""
        if index < 0 or index >= self.size:
            # raise IndexError("Index out of range")
            print(f"索引{index}越界: Index out of range")
            return
        current = self.head
        for _ in range(index):
            current = current.next
        current.value = value
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        values = []
        current = self.head
        while current:
            values.append(str(current.value))
            current = current.next
        return " -> ".join(values)
    
    def __repr__(self):
        return f"MyLinkedList({str(self)})"
    

# 测试手搓双链表
def test_my_doubly_linked_list():
    print("=== 测试手搓双链表 ===")
    
    linked_list = MyLinkedList()
    print(f"初始链表: {linked_list}, 长度: {len(linked_list)}")
    
    # 添加元素
    for i in range(1, 6):
        linked_list.append(i * 10)
    print(f"添加5个元素后: {linked_list}, 长度: {len(linked_list)}")
    
    # 测试 __getitem__
    print(f"\n=== 测试索引访问 ===")
    print(f"linked_list[0] = {linked_list[0]}")
    print(f"linked_list[2] = {linked_list[2]}")
    print(f"linked_list[-1] = {linked_list[-1]}")  # 不支持负数索引
    
    # 测试 __setitem__
    print(f"\n=== 测试修改元素 ===")
    linked_list[0] = 100
    print(f"修改linked_list[0]后: {linked_list}")
    linked_list[4] = 999
    print(f"修改linked_list[4]后: {linked_list}")
    
    # 测试异常处理
    print(f"\n=== 测试异常 ===")
    try:
        print(linked_list[100])
    except IndexError as e:
        print(f"索引100越界: {e}")
    
    try:
        linked_list[100] = 999
    except IndexError as e:
        print(f"索引100越界: {e}")


# 运行测试
if __name__ == "__main__":
    test_my_doubly_linked_list()
````

### 2.4 手搓栈

````python
class MyStack:
    def __init__(self):
        self._stack = []
    
    def push(self, value):
        self._stack.append(value)
    
    def pop(self):
        if not self._stack:
            # raise IndexError("pop from empty stack")
            print("栈空: pop from empty stack")
            return
        return self._stack.pop()
    
    def top(self):
        if not self._stack:
            # raise IndexError("top from empty stack")
            print("栈空: top from empty stack")
            return
        return self._stack[-1]
    
    def __len__(self):
        return len(self._stack)
    
    def __str__(self):
        return str(self._stack)
    
    def __repr__(self):
        return f"MyStack({self._stack})"
    
# 测试手搓栈
def test_my_stack():
    print("=== 测试手搓栈 ===")
    
    stack = MyStack()
    print(f"初始栈: {stack}, 长度: {len(stack)}")
    
    # 压栈
    for i in range(1, 6):
        stack.push(i * 10)
    print(f"压入5个元素后: {stack}, 长度: {len(stack)}")
    
    # 查看栈顶
    print(f"\n=== 测试查看栈顶 ===")
    print(f"栈顶元素: {stack.top()}")
    
    # 弹栈
    print(f"\n=== 测试弹栈 ===")
    popped = stack.pop()
    print(f"弹出元素: {popped}, 弹出后栈: {stack}")
    
    # 测试异常处理
    print(f"\n=== 测试异常 ===")
    empty_stack = MyStack()
    try:
        empty_stack.pop()
    except IndexError as e:
        print(f"弹空栈越界: {e}")
    
    try:
        empty_stack.top()
    except IndexError as e:
        print(f"查看空栈顶越界: {e}")

# 运行测试
if __name__ == "__main__":
    test_my_stack()
````

### 2.5 手搓队列

````python
class MyQueue:
    def __init__(self):
        self._queue = []
    
    def enqueue(self, value):
        self._queue.append(value)
    
    def dequeue(self):
        if not self._queue:
            # raise IndexError("dequeue from empty queue")
            print("队列空: dequeue from empty queue")
            return
        return self._queue.pop(0)
    
    def front(self):
        if not self._queue:
            # raise IndexError("front from empty queue")
            print("队列空: front from empty queue")
            return
        return self._queue[0]
    
    def __len__(self):
        return len(self._queue)
    
    def __str__(self):
        return str(self._queue)
    
    def __repr__(self):
        return f"MyQueue({self._queue})"
    

# 测试手搓队列
def test_my_queue():
    print("=== 测试手搓队列 ===")
    
    queue = MyQueue()
    print(f"初始队列: {queue}, 长度: {len(queue)}")
    
    # 入队
    for i in range(1, 6):
        queue.enqueue(i * 10)
    print(f"入队5个元素后: {queue}, 长度: {len(queue)}")
    
    # 查看队头
    print(f"\n=== 测试查看队头 ===")
    print(f"队头元素: {queue.front()}")
    
    # 出队
    print(f"\n=== 测试出队 ===")
    dequeued = queue.dequeue()
    print(f"出队元素: {dequeued}, 出队后队列: {queue}")
    
    # 测试异常处理
    print(f"\n=== 测试异常 ===")
    empty_queue = MyQueue()
    try:
        empty_queue.dequeue()
    except IndexError as e:
        print(f"出空队列越界: {e}")
    
    try:
        empty_queue.front()
    except IndexError as e:
        print(f"查看空队列头越界: {e}")

# 运行测试
if __name__ == "__main__":
    test_my_queue()
````

### 2.6 手搓二叉树

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class BinaryTree:
    def __init__(self):
        self.root = None
    
    def build_from_list(self, arr):
        """从数组构建二叉树（层次遍历）"""
        if not arr:
            return None
        
        self.root = TreeNode(arr[0])
        queue = [self.root]
        i = 1
        
        while queue and i < len(arr):
            node = queue.pop(0)
            
            # 左子节点
            if i < len(arr) and arr[i] is not None:
                node.left = TreeNode(arr[i])
                queue.append(node.left)
            i += 1
            
            # 右子节点
            if i < len(arr) and arr[i] is not None:
                node.right = TreeNode(arr[i])
                queue.append(node.right)
            i += 1
        
        return self.root
    
    def preorder(self, root=None):
        """前序遍历"""
        if root is None:
            root = self.root
        result = []
        
        def dfs(node):
            if not node:
                return
            result.append(node.val)
            dfs(node.left)
            dfs(node.right)
        
        dfs(root)
        return result
    
    def inorder(self, root=None):
        """中序遍历"""
        if root is None:
            root = self.root
        result = []
        
        def dfs(node):
            if not node:
                return
            dfs(node.left)
            result.append(node.val)
            dfs(node.right)
        
        dfs(root)
        return result
    
    def postorder(self, root=None):
        """后序遍历"""
        if root is None:
            root = self.root
        result = []
        
        def dfs(node):
            if not node:
                return
            dfs(node.left)
            dfs(node.right)
            result.append(node.val)
        
        dfs(root)
        return result
    
    def level_order(self):
        """层次遍历"""
        if not self.root:
            return []
        
        result = []
        queue = [self.root]
        
        while queue:
            node = queue.pop(0)
            result.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        return result
    

# 测试手搓二叉树
def test_my_binary_tree():
    print("=== 测试手搓二叉树 ===")
    
    tree = BinaryTree()
    arr = [1, 2, 3, 4, 5, None, None]
    tree.build_from_list(arr)
    
    print(f"构建的二叉树前序遍历: {tree.preorder()}")
    print(f"构建的二叉树中序遍历: {tree.inorder()}")
    print(f"构建的二叉树后序遍历: {tree.postorder()}")
    print(f"构建的二叉树层次遍历: {tree.level_order()}")

# 运行测试
if __name__ == "__main__":
    test_my_binary_tree()
````

### 2.7 手搓图（邻接表）

````python
class Graph:
    def __init__(self, n, directed=False):
        self.n = n
        self.directed = directed
        self.adj = [[] for _ in range(n)]
    
    def add_edge(self, u, v, weight=1):
        self.adj[u].append((v, weight))
        if not self.directed:
            self.adj[v].append((u, weight))
    
    def bfs(self, start):
        """广度优先搜索"""
        visited = [False] * self.n
        queue = [start]
        visited[start] = True
        result = []
        
        while queue:
            node = queue.pop(0)
            result.append(node)
            
            for neighbor, _ in self.adj[node]:
                if not visited[neighbor]:
                    visited[neighbor] = True
                    queue.append(neighbor)
        
        return result
    
    def dfs(self, start):
        """深度优先搜索"""
        visited = [False] * self.n
        result = []
        
        def _dfs(node):
            visited[node] = True
            result.append(node)
            
            for neighbor, _ in self.adj[node]:
                if not visited[neighbor]:
                    _dfs(neighbor)
        
        _dfs(start)
        return result
    

# 测试手搓图（邻接表）
def test_my_graph():
    print("=== 测试手搓图（邻接表） ===")
    
    graph = Graph(5, directed=True)
    graph.add_edge(0, 1)
    graph.add_edge(0, 2)
    graph.add_edge(1, 3)
    graph.add_edge(2, 3)
    graph.add_edge(3, 4)
    
    print(f"图的邻接表: {graph.adj}")
    print(f"从节点0开始的BFS: {graph.bfs(0)}")
    print(f"从节点0开始的DFS: {graph.dfs(0)}")


# 运行测试
if __name__ == "__main__":
    test_my_graph()
````
